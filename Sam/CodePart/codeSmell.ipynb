## This notebook will be used to read the files generated by Designite tool.

```python
import pandas as pd
```

```python
#read the code smells
codeSmells = pd.read_csv('ImplementationSmells.csv')
```

```python
#get the the types of code smells found in your project
codeSmells['Implementation Smell'].unique().tolist()
```

```python
codeSmells
```

```python
codeSmells.groupby(['Implementation Smell','Type Name']).count()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_tmp = (codeSmells.groupby(['Type Name', 'Implementation Smell']).count().reset_index())

ax = sns.boxenplot(x="Implementation Smell", y="Project Name", data = df_tmp)
ax.set_ylim(0, 15)
plt.xticks(rotation = 90)
```

```python
ax3 = sns.violinplot(x="Implementation Smell", y="Project Name", data = df_tmp)
ax3.set_ylim(0, 15)
plt.xticks(rotation = 90)
```

```python
ax2 = sns.boxplot(x="Implementation Smell", y="Project Name", data = df_tmp)
ax2.set_ylim(0, 15)
plt.xticks(rotation = 90)
```
